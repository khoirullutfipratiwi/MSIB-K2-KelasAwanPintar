```python
berat = int(input("Input berat badan (kg) :"))
tinggi = int(input("Input tinggi badan anda (cm) :"))
BMI = (berat/tinggi*2)

if BMI < 18.5:
    print("Kurang berat badan")
elif BMI >= 18.5:
    print("Normal")
elif BMI >= 23:
    print("Kelebihan berat badan")
elif BMI >= 25:
    print("Obesitas tingkat 1")
elif BMI >= 30:
    print("Obesitas tingkat 2")
else:
    print("Tidak masuk manapun")
```

2. Buatlah sebuah program yang akan terus looping sampai user menginput angka nol! Setelah user menginput angka nol, program kemudian akan menghitung rata-rata seluruh angka yang telah diinput oleh user sebelumnya. Contoh output program:

soal flowcontrol 1.PNG


```python
number_list = []

while True:
    number = input("Masukkan angka : ")
    if number.isdigit():
        number = int(number)
        if number == 0:
            break
        number_list.append(number)
    else:
        print("Error!")

if len(number_list) > 0:
    rata_rata = sum(number_list) / len(number_list)
    print(f"Rata-rata semua angka yang telah diinput sebelumnya adalah  {rata_rata}")
```

3. Buatlah program yang dapat memisahkan nilai berikut ke dalam list yang sesuai!

soal flowcontrol 22 (1).png

**Program harus mengimplementasikan conditionals dan loops!**

```python
nilai = [67, 46, 52, -12, 50, 73, 11, 61, 101, 41, 86, 86, 74, 88, 43, 67, 71, 47, 81, 45, 85, 17, -1, 39, 20, 79]

grades = {
    "Grade A": [],
    "Grade B": [],
    "Grade C": [],
    "Grade D": [],
    "Grade E": [],
    "Nilai tidak valid": []
}

for n in nilai:
    if 80 <= n <= 100:
        grades["Grade A"].append(n)
    elif 70 <= n < 80:
        grades["Grade B"].append(n)
    elif 50 <= n < 70:
        grades["Grade C"].append(n)
    elif 40 <= n < 50:
        grades["Grade D"].append(n)
    elif 0 <= n < 40:
        grades["Grade E"].append(n)
    else:
        grades["Nilai tidak valid"].append(n)

print("Grade A\t\t  :", grades["Grade A"])
print("Grade B\t\t  :", grades["Grade B"])
print("Grade C\t\t  :", grades["Grade C"])
print("Grade D\t\t  :", grades["Grade D"])
print("Grade E\t\t  :", grades["Grade E"])
print("Nilai tidak valid :", grades["Nilai tidak valid"])
```

### 1. Program Penghitung Upah Karyawan



Sebuah perusahaan meminta Anda membuat sebuah program untuk menghitung upah bersih karyawan mereka. Berikut adalah sistem pengupahan di perusahaan tersebut:

pengupahan (1).png


Program harus mengimplementasikan konsep function yang memanggil function lainnya (*divide and conquer*)! Jangan lupa tambahkan dokumentasi di dalam function menggunakan docstring!

Lalu, hitunglah upah bersih karyawan berikut ini menggunakan program yang telah dibuat:

karyawan (2).png

```python
karyawan_dict = {
    'Anna' : [2, 10],
    'Dimas' : [3, 15],
    'Sriyono' : [1, 3],
    'Oky' : [2, 8],
    'Wati' : [1, 6],
    'Yayan' : [3, 17]
}

def golongan(gol):
    if gol == 1:
        gaji_pokok = 15_000_000
        tunjangan_persen = 0.15
        upah_overtime = 0
        tunjangan = tunjangan_persen * gaji_pokok

    elif gol == 2:
        gaji_pokok = 7_500_000
        tunjangan_persen = 0.10
        upah_overtime = 150_000
        tunjangan = tunjangan_persen * gaji_pokok

    elif gol == 3:
        gaji_pokok = 5_000_000
        tunjangan_persen = 0.05
        upah_overtime = 100_000
        tunjangan = tunjangan_persen * gaji_pokok

    else:
        gaji_pokok = 0
        tunjangan_persen = 0
        upah_overtime = 0
        tunjangan = tunjangan_persen * gaji_pokok

    return gaji_pokok, upah_overtime, tunjangan

for key, value in karyawan_dict.items():
    gol = value[0]
    ovt = value[1]
    data = golongan(gol)
    upah_kotor = data[0] + (data[1] * ovt) + data[2]
    potongan = 0.02 * upah_kotor
    upah_bersih = upah_kotor - potongan
    print(f"Nama Karyawan : {key} \nUpah kotor    : Rp {upah_kotor} => mendapat potongan Rp {potongan} \nUpah bersih   : Rp {upah_bersih}\n")
```
